```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=128
t = np.linspace(0, 2*np.pi, Nt)

z0=10
X, Y, Z = np.cos(t), np.sin(t), z0

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=10**4
t = np.linspace(0, 1*np.pi, Nt)

z0, t0 =10, 10
X, Y, Z = (t**2 - t0**2)/(t**2 + t0**2), 2*t*t0/(t**2 + t0**2), z0

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

Nt=128
t = np.linspace(0, 3.6*np.pi, Nt)

vo=10
X, Y, Z = np.cos(t), np.sin(t), vo*t

sp=ax.scatter3D(X, Y, Z, c=t)
fig.colorbar(sp);
```
